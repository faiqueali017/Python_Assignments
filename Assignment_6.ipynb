```python
#Question: 5

class Car:
    
    def __init__(self,carName,carModel,carMake,carColor):
        self.carName = carName
        self.carModel = carModel
        self.carMake = carMake
        self.carColor = carColor
        
    def carDetails(self):
        details = "\nCar Name: {} \nCar Model: {} \nCar Make: {} \nCar Color: {} ".format(
            self.carName,self.carModel,self.carMake,self.carColor)
        return details
    

myCar1 = Car("Civic","2019","Honda","Black")

myCar2 = Car("Corolla","2017","Toyota","White")

myCar3 = Car("AudiR8","2019","Audi","Black")

myCar4 = Car("Cultus","2015","Suzuki","Silver")

myCar5 = Car("Vitz","2014","Toyota","Grey")

print(myCar1.carDetails())
print(myCar2.carDetails())
print(myCar3.carDetails())
print(myCar4.carDetails())
print(myCar5.carDetails())
```
